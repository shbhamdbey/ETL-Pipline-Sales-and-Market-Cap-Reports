# Importing Libraries

```python
import requests
from bs4 import BeautifulSoup
```

```python
r = requests.get(url)
```

# Capturing HTML file format

```python
soup = BeautifulSoup(r.text, 'html.parser')
```

```python
tab =[]
mark =[]
```

```python
picktable = soup.find('table', class_ ='scroll responsive display nowrap data-t data-t wpDataTable wpDataTableID-6012')
```

```python
martable = soup.find('table', class_ ='responsive display nowrap data-t data-t wpDataTable wpDataTableID-6047')
```

```python
print(martable)
```

# Mapping the Required Feilds

```python
for team in picktable.find_all('tbody'):
    rows = team.find_all('tr')
    for row in rows:
        tab.append({
        'year': row.find('td').text,
        'jan': row.find_all('td')[1].text,
        'feb': row.find_all('td')[2].text,
        'mar': row.find_all('td')[3].text,
        'apr': row.find_all('td')[4].text,
        'may': row.find_all('td')[5].text,
        'jun': row.find_all('td')[6].text,
        'jul': row.find_all('td')[7].text,
        'aug': row.find_all('td')[8].text,
        'sep': row.find_all('td')[9].text,
        'oct': row.find_all('td')[10].text,
        'nov': row.find_all('td')[11].text,
        'decb': row.find_all('td')[12].text
        })
        print(tab)
```

```python
for team in martable.find_all('tbody'):
    rows = team.find_all('tr')
    for row in rows:
        mark.append({
        'year': row.find('td').text,
        'Sales': row.find_all('td')[1].text,
        'YOY_Change': row.find_all('td')[2].text,
        'US_Mark': row.find_all('td')[3].text,
        'Mark_Share': row.find_all('td')[4].text

        })
        print(mark)
```
